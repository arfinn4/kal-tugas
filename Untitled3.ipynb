```python
import matplotlib.pyplot as plt

# Titik segitiga (objek)
points = [[-5, 1], [-1, 4], [-2, -1]]
drag_index = None

def reflect(points):
    # Refleksi sumbu-Y
    return [[-x, y] for x, y in points]

def draw():
    plt.cla()

    # Objek asli
    x = [p[0] for p in points] + [points[0][0]]
    y = [p[1] for p in points] + [points[0][1]]
    plt.plot(x, y, 'ro-', label='Objek')

    # Bayangan (selalu dihitung ulang)
    ref = reflect(points)
    xr = [p[0] for p in ref] + [ref[0][0]]
    yr = [p[1] for p in ref] + [ref[0][1]]
    plt.plot(xr, yr, 'bo-', label='Bayangan')

    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
    plt.xlim(-6, 6)
    plt.ylim(-6, 6)
    plt.grid(True)
    plt.legend()
    plt.draw()

def on_click(event):
    global drag_index
    if event.xdata is None: return

    # pilih titik terdekat
    for i, (x, y) in enumerate(points):
        if abs(x - event.xdata) < 0.3 and abs(y - event.ydata) < 0.3:
            drag_index = i
            break

def on_release(event):
    global drag_index
    drag_index = None

def on_move(event):
    if drag_index is None or event.xdata is None:
        return

    # geser titik
    points[drag_index] = [event.xdata, event.ydata]

    # redraw → bayangan otomatis ikut
    draw()

fig = plt.figure()
draw()

fig.canvas.mpl_connect('button_press_event', on_click)
fig.canvas.mpl_connect('button_release_event', on_release)
fig.canvas.mpl_connect('motion_notify_event', on_move)

plt.show()
```
